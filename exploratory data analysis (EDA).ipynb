## Load the data

Load the "Titanic Dataset.csv" file into a pandas DataFrame.


```python
import pandas as pd

df = pd.read_csv('/content/Titanic Dataset.csv')
display(df.head())
```

**Reasoning**:
Display the first few rows, check for missing values, and look at the data types of each column as requested in the instructions.



```python
display(df.head())
display(df.info())
display(df.isnull().sum())
```

## Handle missing values

Address any missing values found in the dataset using appropriate techniques (e.g., imputation or removal).


```python
# Drop columns with a large number of missing values
df = df.drop(['cabin', 'body', 'home.dest'], axis=1)

# Impute missing values in 'age' with the median
df['age'].fillna(df['age'].median(), inplace=True)

# Impute missing values in 'fare' with the median
df['fare'].fillna(df['fare'].median(), inplace=True)

# Impute missing values in 'embarked' with the mode
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

# Verify that missing values have been handled
display(df.isnull().sum())
```

**Reasoning**:
The missing values in 'age', 'fare', and 'embarked' have been handled. The 'boat' column still has a large number of missing values, which was not dropped in the previous step. Based on the instructions, this column should be considered for dropping due to the high number of missing values. After dropping this column, all the missing values will be handled, and the subtask will be completed.



```python
# Drop the 'boat' column due to a large number of missing values
df = df.drop(['boat'], axis=1)

# Verify that missing values have been handled
display(df.isnull().sum())
```

## Clean and transform data

Perform necessary data cleaning and transformations (e.g., convert data types, handle outliers).


```python
display(df.info())
df['survived'] = df['survived'].astype('category')
display(df.info())
```

**Reasoning**:
Visualize the distributions of the numerical columns to identify potential outliers.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'fare', 'sibsp', 'parch']
plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Box plot of {col}')
plt.tight_layout()
plt.show()
```

**Reasoning**:
Based on the box plots, there are outliers in 'age', 'fare', 'sibsp', and 'parch'. Remove the outliers in 'fare' by keeping values within 3 standard deviations from the mean. For 'age', 'sibsp', and 'parch', decide to keep the outliers as they might represent genuine cases and removing them could lead to loss of valuable information.



```python
# Remove outliers in 'fare' using the 3 standard deviation rule
mean_fare = df['fare'].mean()
std_fare = df['fare'].std()
df = df[(df['fare'] > mean_fare - 3 * std_fare) & (df['fare'] < mean_fare + 3 * std_fare)]

# Display box plot of 'fare' after outlier removal
plt.figure(figsize=(6, 4))
sns.boxplot(y=df['fare'])
plt.title('Box plot of fare after outlier removal')
plt.show()
```

## Univariate analysis


```python
numerical_cols = ['age', 'fare', 'sibsp', 'parch']
categorical_cols = ['pclass', 'survived', 'sex', 'embarked']

print("Descriptive statistics for numerical columns:")
display(df[numerical_cols].describe())

print("\nValue counts for categorical columns:")
for col in categorical_cols:
    print(f"\nValue counts for {col}:")
    display(df[col].value_counts())
```

**Reasoning**:
Create histograms for numerical columns and count plots for categorical columns to visualize their distributions as requested in the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['age', 'fare', 'sibsp', 'parch']
categorical_cols = ['pclass', 'survived', 'sex', 'embarked']

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols):
    plt.subplot(2, 2, i + 1)
    sns.histplot(data=df, x=col, kde=True)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 8))
for i, col in enumerate(categorical_cols):
    plt.subplot(2, 2, i + 1)
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')

plt.tight_layout()
plt.show()
```

## Bivariate analysis
 the relationships between pairs of variables to identify patterns and trends.


```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create a correlation matrix for the numerical columns and display it using a heatmap.
numerical_cols = ['age', 'fare', 'sibsp', 'parch']
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()

# 2. Generate box plots to visualize the relationship between categorical and numerical variables.
categorical_cols = ['pclass', 'survived', 'sex', 'embarked']
numerical_cols_for_plotting = ['age', 'fare']

for cat_col in categorical_cols:
    for num_col in numerical_cols_for_plotting:
        plt.figure(figsize=(8, 6))
        sns.boxplot(data=df, x=cat_col, y=num_col)
        plt.title(f'{num_col} by {cat_col}')
        plt.show()


# 3. Create scatter plots to explore the relationship between pairs of numerical variables.
sns.pairplot(df[numerical_cols])
plt.suptitle('Scatter Plots of Numerical Variables', y=1.02)
plt.show()


# 4. Generate grouped bar plots to visualize the relationship between two categorical variables.
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='pclass', hue='survived')
plt.title('Survival Count by Pclass')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='sex', hue='survived')
plt.title('Survival Count by Sex')
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='embarked', hue='survived')
plt.title('Survival Count by Embarked')
plt.show()
```

## Multivariate analysis

### Subtask:
Explore relationships among multiple variables to gain deeper insights.


**Reasoning**:
Generate plots to explore relationships among multiple variables as requested in the instructions.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Create a grouped box plot to visualize the distribution of 'fare' across different 'pclass' and 'survived' categories.
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='pclass', y='fare', hue='survived')
plt.title('Distribution of Fare by Pclass and Survival')
plt.show()

# 2. Generate a facet grid or similar plot to explore the relationship between 'survived', 'pclass', and 'sex'.
g = sns.FacetGrid(df, col='pclass', row='survived', margin_titles=True)
g.map_dataframe(sns.countplot, x='sex')
g.set_titles(row_template='Survived: {row_name}', col_template='Pclass: {col_name}')
g.fig.suptitle('Survival by Pclass and Sex', y=1.03)
g.tight_layout()
plt.show()

# 3. Create a scatter plot with color-coded points based on 'survived' to visualize the relationship between 'age' and 'fare'.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='age', y='fare', hue='survived', alpha=0.6)
plt.title('Relationship between Age and Fare by Survival')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualize the relationship between Age and Survival using a violin plot
plt.figure(figsize=(8, 6))
sns.violinplot(data=df, x='survived', y='age')
plt.title('Distribution of Age by Survival')
plt.show()

# Visualize the relationship between Fare and Survival using a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, x='survived', y='fare')
plt.title('Distribution of Fare by Survival')
plt.show()
```
